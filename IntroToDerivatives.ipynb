## Example 1
Find the equation for the tangent line of $\Large f(x) = \frac{3}{x}$ at $(2, 3/2)$

Answer: $\Large l(x) = -\frac{3}{4}x + 3$

```python
f = 3/x
l = -3*x/4 + 3

p1 = plot(f, (x, 1, 3))
p2 = plot(l, (x, 1, 3), color='red')
p3 = circle((2, 3/2), 0.05, rgbcolor=(0, 0, 1))
show(p1+p2+p3)
```

## Example 2
Find the equation for the tangent line of $\Large f(x) = \sqrt{x}$ at $(25, 5)$

Answer: $\Large l(x) = \frac{x}{10} + \frac{9}{2}$

```python
f = x^(1/2)
l = x/10 + 5/2

p1 = plot(f, (x, 0, 60), aspect_ratio=8)
p2 = plot(l, (x, 0, 60), color='red', aspect_ratio='automatic')
p3 = circle((25, 5), 0.5, rgbcolor=(0, 0, 1))
show(p1+p2+p3)
```

## Example 3
Find the equation for the tangent line of $\Large f(x) = x^3 + 4x$ at $(1, 5)$

Answer: $\Large l(x) = 7x - 2$

```python
f = x^3 + 4*x
l = 7*x - 2

p1 = plot(f, (x, 0, 20), ymin=0, ymax=10)
p2 = plot(l, (x, 0, 20), ymin=0, ymax=10, color='red')
p3 = circle((1, 5), 0.5, rgbcolor=(0, 0, 1))
show(p1+p2+p3, figsize=(10, 3))
```

## Example 4

$\Large f(x) = x^2$


$\Large f'(x) = 2x$


```python
f = x^2+2
fprime = 2*x
p1 = plot(f, (x, -3, 3))
p2 = plot(fprime, (x, -3, 3), color='red')
show(p1+p2)
```

## Example 5

$\Large f(x) = |x| = \left\{  \begin{array}{cc} x & x > 0 \\ -x & x \leq 0\end{array} \right\}$

$\Large f'(x) = \left\{  \begin{array}{cc} 1 & x > 0 \\ -1 & x < 0\end{array} \right\}$

```python
f = lambda x: abs(x)

def fprime(x):
    if x > 0:
        return 1
    if x < 0:
        return -1
    return nan
p1 = plot(f, (x, -2, 2))
p2 = plot(fprime, (x, -2, 2), color='red')
show(p1 + p2)
```
